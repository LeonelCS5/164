```python
# Import the necessary modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LogisticRegression


# Read in the file fg_attempt.csv
iris_dataset = pd.read_csv('iris_data.csv')

print (f"Instances and Features: {iris_dataset.shape}")
x1Name = 'petal_length'
x2Name = 'petal_width'
x3Name = 'sepal_length'
x4Name = 'sepal_width'
# Create a dataframe X containing 2 features
X = iris_dataset[[x1Name, x2Name]]
# Create a dataframe y containing 'class'
y = iris_dataset[['class']]

# Flatten y into an array
yArray = np.ravel(y)

# Initialize a LogisticRegression() model
logisticModel = LogisticRegression()

# Fit the model
logisticModel.fit(X, yArray)

# Input feature values for a sample instance
x1 = float(input("Enter petal length: "))
x2 = float(input("Enter petal width:  "))

# Create a new dataframe with user-input
XNew = pd.DataFrame({x1Name: [x1], x2Name: [x2]})

# Predict the outcome from the new data
pred = logisticModel.predict(XNew)
print(f"prediction: {pred}")

# Determine the accuracy of the model logisticModel
score = logisticModel.score(X, yArray)
print(f"accuracy: {score:.2f}")
```

Scatter Plot of Iris Dataset:
petal length & petal with VS sepal length & sepal width.

```python
p = sns.scatterplot(data=iris_dataset, x =x1Name, y=x2Name, hue='class', style='class')
p.set_xlabel('Petal Length', fontsize=14)
p.set_ylabel('Petal Width', fontsize=14)
p.set_title("Petal length & Petal width")
```

```python
p = sns.scatterplot(data=iris_dataset, x =x3Name, y=x4Name, hue='class', style='class')
p.set_xlabel('Sepal Length', fontsize=14)
p.set_ylabel('Sepal Width', fontsize=14)
p.set_title("Sepal length & Sepal width")
```
